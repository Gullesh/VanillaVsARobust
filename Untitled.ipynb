```python
import torch
from torch.autograd import Variable
import cv2
import sys
import numpy as np
import time
import torchvision 
import torchvision.transforms as transforms
import torchray.benchmark
import torchvision.models as models
from matplotlib import pyplot as plt
from torchray.attribution.extremal_perturbation import extremal_perturbation, contrastive_reward, simple_reward
torch.cuda.empty_cache()
```

```python
    
val_transforms = transforms.Compose([transforms.Resize(256),transforms.CenterCrop(224), transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])])    
vall = torchray.benchmark.datasets.ImageFolder('/home/mallet/Desktop/VanillaVsARobust/validationSample',transform = val_transforms)
```

```python
use_cuda = torch.cuda.is_available()
def load_model():
    model = models.vgg19(pretrained=True)
    model.eval()
    if use_cuda:
        model.cuda()

    for p in model.features.parameters():
        p.requires_grad = False
    for p in model.classifier.parameters():
        p.requires_grad = False

    return model

img, labele = vall[1403]                               
print(img.shape)
x = img.unsqueeze(0)
x = x.cuda()
print(x.shape)
```

```python
masks_1, _ = extremal_perturbation(
    load_model(), x, labele,
    reward_func=simple_reward,
    debug=True,
    areas=[0.12])
print(masks_1.shape)
```

```python

print(masks_1)
torch.max(masks_1)
```

```python
saliency = torch.squeeze(masks_1)
print(saliency.shape)
saliency = saliency.cpu().detach().numpy()
plt.imshow(saliency, interpolation='nearest')
plt.show()
```

```python
imag = 1-saliency
imag.max()
plt.imshow(imag, interpolation='nearest')
plt.show()
```
